```python
print('Please enter the origin:')
x0 = int(input('x0: '))
y0 = int(input('y0: '))
a = int(input('Please enter a: '))
b = int(input('Please enter b (b<a): '))
lowlim = x0 - a/2
uplim = x0 + a/2
x = np.arange(lowlim, uplim, 50)
y = b*(math.sqrt(1 - (x**2 / a**2)))
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
```

гиперболы.

```python
k = int(input('Please enter k:'))
x1 = []
x2 = []
y1 = []
y2 = []
for i in range(-100, 0):
    x1_ = i/100
    x2_ = -i/100
    x1.append(x1_)
    x2.append(x2_)
    y1.append(k / x1_)
    y2.append(k / x2_)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.xlabel("x")
plt.ylabel("y")
```

5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-20, 20, 2)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 3*Y
Z2 = 2*X + 3*Y + 15
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
# ax.scatter(0, 0, 0, 'O', 100,'red')
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.2)
Y = np.arange(-5, 5, 0.2)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**2 + Y**2)
Z1 = np.sqrt(X**2 - Y**2)
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z1)
show()
```
