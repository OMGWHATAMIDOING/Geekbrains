4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-3, 12, 100)
plt.plot(x, (1 - (1 - np.exp(x))/x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-3,4) 
plt.ylim(-2,10)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return(y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)
x1, y1 = fsolve(equations, (-2, 2))
x2, y2 = fsolve(equations, (2, 4))
print(x1, y1, x2, y2)
```
